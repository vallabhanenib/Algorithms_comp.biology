```python
import pandas as pd
  
# read text file into pandas DataFrame
GeneIDlist = pd.read_table("gene_annotation.txt", delimiter='\t')
  
# display DataFrame
print(GeneIDlist)
```

```python
Unique_GeneID = GeneIDlist['GeneID'].unique()
print('Unique elements in column "GeneID" ')
print(Unique_GeneID)

Unique_GeneIDCount = GeneIDlist['GeneID'].nunique()
print('Number of unique values in column "GeneID" of the dataframe : ')
print(Unique_GeneIDCount)
```
